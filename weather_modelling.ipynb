```python
import matplotlib.pyplot as plt
```

```python
def quadratic_t_model(day,a,b,c):
  temp=a*(day**2)+b*day+c
  return temp
```

```python
def main():
  a,b,c=1,-3,2
  num_days=int(input("enter the number of days to model:"))
  days=list(range(1,num_days+1))
  temp=[quadratic_t_model(day,a,b,c) for day in days]
  plt.plot(days,temp,label=f'Tempearture:{a}x^2+{b}x+{c}')
  plt.title('Weather Modelling')
  plt.xlabel('Day')
  plt.ylabel('Temperature(c)')
  plt.legend()
  plt.show()
main()
```

```python
def main():

    a = float(input("Enter coefficient a (for x² term): "))
    b = float(input("Enter coefficient b (for x term): "))
    c = float(input("Enter coefficient c (constant term): "))


    num_days = int(input("Enter the number of days to model: "))

    days = list(range(1, num_days + 1))

    temp = [quadratic_t_model(day, a, b, c) for day in days]


    plt.plot(days, temp, marker='o', label=f'Temperature: {a}x² + {b}x + {c}')
    plt.title('Weather Modelling')
    plt.xlabel('Day')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    plt.grid(True)
    plt.show()

main()
```

```python
import pandas as pd
```

```python
def quadratic_weather_model(time,a,b,c):
  temperature=a*(time**2)+b*time+c
  return temperature
```

```python
def main():
    print("Quadratic Weather Modeling")
    print("==========================")

    try:
        data = """a,b,c
                 1,-3,2
                 """
        with open("wea2.csv", "w") as f:
          f.write(data)
        df = pd.read_csv("wea2.csv")
        time_values = list(range(0, 11))
        plt.figure(figsize=(8, 6))

        for index, row in df.iterrows():
            x, y, z = row['a'], row['b'], row['c']
            temperature_values = [quadratic_weather_model(t, x, y, z) for t in time_values]
            plt.plot(time_values, temperature_values, marker='o', linestyle='--',
                     label=f'a={x}, b={y}, c={z}')

        plt.title('Temperature Variation Over Time')
        plt.xlabel('Time')
        plt.ylabel('Temperature')
        plt.grid(True)
        plt.xlim(0, 10)
        plt.legend()
        plt.show()

    except FileNotFoundError:
        print("File not found. Please make sure 'wea2.csv' exists.")

main()
```
